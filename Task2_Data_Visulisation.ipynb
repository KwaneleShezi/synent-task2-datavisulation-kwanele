```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the iris dataset
df = pd.read_csv('Iris.csv')
print(df.shape)
df.head()
```

```python
# check column names and data types
print(df.columns)
print(df.info())
```

```python
# count how many flowers are in each species and plot a bar chart
df['Species'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Flowers per Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# show the distribution of petal length across all flowers
plt.hist(df['PetalLengthCm'], bins=20, color='orange')
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# compare sepal length vs petal length for each species
for species in df['Species'].unique():
    data = df[df['Species'] == species]
    plt.scatter(data['SepalLengthCm'], data['PetalLengthCm'], label=species)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
